```python
import csv
csvfile=open('Climate.csv')
reader=csv.reader(csvfile)
a=[]
num_attributes=6
```

```python
for row in reader:
  a.append(row)
  print(row)
csvfile.close()
```

```python
hypothesis=['0']*num_attributes
print(hypothesis)
```

```python
for j in range(0,num_attributes):
    hypothesis[j]=a[1][j]
```

```python
for i in range(0,len(a)):
    if a[i][num_attributes]=='Yes':
        for j in range(0,num_attributes):
            if a[i][j]!=hypothesis[j]:
                hypothesis[j]='?'
            else:
                hypothesis[j]=a[i][j]
    print("  the hypothesis: {0}".format(i),hypothesis)
```
